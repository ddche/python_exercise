```python
# exercise 25
# split()  字符串切片（string切片存为list）
def break_words(stuff):
    """the function will break up words for us."""
    words = stuff.split(" ")
    return words
# sorted()   string分类(对单词按照字母排序；对单词组 按照 单词首字母排序)
def sort_words(words):
    """sorts the words."""
    return sorted(words)
# pop()   list选取并删除该项
def print_first_word(words):
    """prints the first word after popping it off."""
    word = words.pop(0)
    print(word)
def print_last_word(words):
    """prints the last word after popping it off"""
    word = words.pop(-1)
    print(word)
def sort_sentence(sentence):
    """take in a full sentence and returns the sorted words."""
    words = break_words(sentence)
    print(words)
    return sort_words(words)
def print_first_and_last(sentence):
    """prints the first and last words of the sentence."""
    words = break_words(sentence)
    first_words = print_first_word(words)
    last_words = print_last_word(words)
    return first_words,last_words
def print_first_and_last_sorted(sentence):
    """sorts the words then prints the first and last one"""
    words = sentence.split(" ")
    first_word = sorted(words).pop(0)
    last_word = sorted(words).pop(-1)
    return first_word,last_word
       
sarcasm=["sarcasm","ambiguity","negation"]
print(break_words("neutral"))
print(break_words("neutral negation sarcasm ambiguity"))
print(sort_words("netural negation sarcasm ambiguity"))
# print(sort_words("netural"))
print(print_first_word(sarcasm))
print(print_last_word(sarcasm))
print(sort_sentence("netural negation sarcasm ambiguity"))
print(print_first_and_last("netural negation sarcasm ambiguity"))
print(print_first_and_last_sorted("netural negation sarcasm ambiguity"))
```
